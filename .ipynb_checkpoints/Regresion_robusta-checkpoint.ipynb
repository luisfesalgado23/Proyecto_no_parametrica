```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python
def huber_loss(u, c):
    return np.where(np.abs(u) <= c, 0.5 * u**2, c * (np.abs(u) - 0.5 * c))

def tukey_bisquare_weights(u, k):
    return np.where(np.abs(u) <= k, (1 - (u / k)**2)**2, 0)

def loss_function(beta, X, y, loss_fn, weights):
    residuals = y - X.dot(beta)
    return np.sum(weights * loss_fn(residuals, 1))

def fit_m_estimator(X, y, loss_function, weights_fn, c, k, max_iterations=100, tol=1e-4):
    n, p = X.shape
    beta = np.zeros(p)

    result = minimize(lambda b: loss_function(b, X, y, huber_loss, weights_fn(X.dot(b) - y, k)),
                      beta, method='L-BFGS-B', options={'disp': True, 'maxiter': max_iterations, 'gtol': tol})
    
    return result.x
```

```python
def l1_loss(u):
    return np.sum(np.abs(u))

def l1_regression_loss(beta, X, y):
    residuals = y - X.dot(beta)
    return l1_loss(residuals)

def l1_regression(X, y, max_iterations=100, tol=1e-4):
    n, p = X.shape
    beta = np.zeros(p)

    result = minimize(lambda b: l1_regression_loss(b, X, y), beta, method='L-BFGS-B', options={'disp': True, 'maxiter': max_iterations, 'gtol': tol})
    
    return result.x
```

```python
# Configurar la semilla para reproducibilidad
np.random.seed(42)

# Generar datos sin atípicos
X = np.random.rand(100, 1)
y = 2 * X.squeeze() + 1 + 0.1 * np.random.randn(100)

# Introducir más valores atípicos
num_outliers = 20
outlier_indices = np.random.choice(100, num_outliers, replace=False)
y[outlier_indices] += 10  # Aumentar el valor de los outliers

# Aplicar regresión de Mínimos Cuadrados Medianos
X_with_bias = np.column_stack((np.ones_like(X), X))  # Agregar término de sesgo
betas_l1_regression = l1_regression(X_with_bias, y)

# Visualizar los resultados
plt.scatter(X, y, label='Datos con Outliers')
plt.plot(X, X_with_bias.dot(betas_l1_regression), color='red', label='Regresión L1')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

# Mostrar resultados
print("Coeficientes L1 regression:", betas_l1_regression)
```

```python
# Configurar la semilla para reproducibilidad
np.random.seed(42)

# Generar datos sin atípicos
X = np.random.rand(100, 1)
y = 2 * X.squeeze() + 1 + 0.1 * np.random.randn(100)

# Introducir más valores atípicos
num_outliers = 20
outlier_indices = np.random.choice(100, num_outliers, replace=False)
y[outlier_indices] += 10  # Aumentar el valor de los outliers

# Ajustar el estimador M a los datos simulados
c = 1.345  # Parámetro de la función de pérdida Huber
k = 4.685  # Parámetro de la función de pesos bisquare de Tukey

# Llamada a la función fit_m_estimator
coefficients = fit_m_estimator(X, y, loss_function, tukey_bisquare_weights, c, k)

# Visualizar los resultados
plt.scatter(X, y, label='Datos con atípico')
plt.plot(X, X.dot(coefficients), color='red', label='Ajuste robusto')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
#Calcular la pérdida de Huber en los datos de entrenamiento
huber_loss_values = huber_loss(y - X.dot(coefficients), c)

# Calcular el valor total de la función de pérdida de Huber
total_huber_loss = np.sum(huber_loss_values)

print(f"Pérdida total de Huber en los datos de entrenamiento: {total_huber_loss}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Ajustar regresión lineal ordinaria
linear_model = LinearRegression()
linear_model.fit(X, y)
y_pred_linear = linear_model.predict(X)

# Calcular el MSE para la regresión lineal
mse_linear = mean_squared_error(y, y_pred_linear)

print(f"MSE para regresión lineal ordinaria: {mse_linear}")

# Visualizar los resultados
plt.scatter(X, y, label='Datos con atípico')
plt.plot(X, y_pred_linear, color='green', label='Regresión lineal ordinaria')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Visualizar los resultados y residuos
plt.scatter(X, y, label='Datos con atípico')
plt.plot(X, y_pred_linear, color='green', label='Regresión lineal ordinaria')
plt.plot(X, X.dot(coefficients), color='red', label='Ajuste robusto (Estimador M)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Comparación de Regresiones')
plt.show()

# Calcular y comparar MSE
mse_linear = mean_squared_error(y, y_pred_linear)
mse_robust = mean_squared_error(y, X.dot(coefficients))

print(f"MSE para regresión lineal ordinaria: {mse_linear}")
print(f"MSE para ajuste robusto (Estimador M): {mse_robust}")
```
